```python
training = open('./TrainingData.csv', encoding='utf8').read().strip().split('\n')
validation = open('./ValidationData.csv', encoding='utf8').read().strip().split('\n')

labels = training[0].split(',')

t = []
v = []
for row in training[1:]:
  sp = row.split(',')
  for i in range(len(sp)):
    if i != 520 and i != 521:
      sp[i] = int(sp[i])
    else:
      sp[i] = float(sp[i])
  t.append(sp)

for row in validation[1:]:
  sp = row.split(',')
  for i in range(len(sp)):
    if i != 520 and i != 521:
      sp[i] = int(sp[i])
    else:
      sp[i] = float(sp[i])
  v.append(sp)

print(labels)
print(t[0])
print(v[0])
```

```python
# Build X
# X : [WAP001, ..., WAP520, floor, buildingId, spaceId,
#     relativePos, UserId, PhoneId, Timestamp]

x_labels = labels[:520] + labels[522:]
x_t = [x[:520] + x[522:] for x in t]
x_v = [x[:520] + x[522:] for x in v]

print(x_labels[520:])
print(x_t[0][520:])
print(x_v[0][520:])
```

```python
# Build Y
# Y : [Lat, Lon]

y_labels = labels[520:522]
y_t = [x[520:522] for x in t]
y_v = [x[520:522] for x in v]

print(y_labels)
print(y_t[0])
print(y_v[0])
```
